```python
def FTM_EKF(df, df_idx, X_update, P_update, ap_locs):

    # if ZUPT_flag == 1:
        # print("time exact match", df.time[df_idx])
        # Access point SSIDs
    ssids = ['  portal1 ', '  portal2 ', '  portal3 ', '  portal4 ', '  portal5 ', '  portal6 ', '  portal7 ']

    d = np.ones([4,1])*-1

    rssi = np.ones([4,1])

    ap_locs_rec = np.ones([4,2])*-1

    r_0 = np.ones([4,1])*-1

    H = np.zeros([4,2])

    A = np.eye(2)
    
    X_prev = X_update[0:2].copy()
    X_prev = X_prev.reshape([2,1])
#     print(f"{X_prev.shape}")
    
    # Measurement noise covariance matrix
    R = np.eye(4)*8e3
    # Process noise covariance matrix
    Q = np.eye(2)*5

    Z = np.zeros_like(d)
# 
    P_prev = P_update.copy()
# 
    # get distances and rssi from dataframe
    d[0,0],d[1,0],d[2,0],d[3,0] = df.AP1_DIST[df_idx],df.AP2_DIST[df_idx],df.AP3_DIST[df_idx],df.AP4_DIST[df_idx]
    rssi[0,0],rssi[1,0],rssi[2,0],rssi[3,0] = df.AP1_RSSI[df_idx],df.AP2_RSSI[df_idx],df.AP3_RSSI[df_idx],df.AP4_RSSI[df_idx]

    # get locations of received access points 
    ap_locs_rec[0],ap_locs_rec[1],ap_locs_rec[2],ap_locs_rec[3] = ap_locs[ssids.index(df.AP1[df_idx]),0:2],ap_locs[ssids.index(df.AP2[df_idx]),0:2],ap_locs[ssids.index(df.AP3[df_idx]),0:2],ap_locs[ssids.index(df.AP4[df_idx]),0:2]

    # Measurement prediction and Jacobian calculation
    for j in range(4):
        temp = np.square(ap_locs_rec[j,0] - X_prev[0,0]) + np.square(ap_locs_rec[j,1] - X_prev[1,0])
        # temp = np.square(ap_locs_rec[j,0] - gt[0]) + np.square(ap_locs_rec[j,1] - gt[1])
        r_0[j,0] = np.sqrt(temp)

    for j in range(4):
        for k in range(2):
            if (r_0[j,0] == 0):
                H[j,k] = 0
            else:
                H[j,k] = (X_prev[k,0] - ap_locs_rec[j,k])/r_0[j,0]

    # Implement correction from BNO055 ZUPT update
    
    # Kalman gain calculation
    K = np.linalg.inv((H @ P_prev @ H.T) + R)

    K = P_prev @ H.T @ K

    # Measurement update and State propagation
    for j in range(4):
        Z[j,0] = d[j,0] - r_0[j,0]

    error = K @ Z

    X_next = A @ (X_prev + error)
#     print(f"{X_next.shape}")

    P_next = (np.eye(2) - K @ H) @ P_prev

    X_update = np.array([X_next[0,0],X_next[1,0],X_update[2]])
    P_update = (A @ P_next @ A.T) + Q
#     print(P_update)
    return X_update, P_update
```

```python
def IMU_EKF(accx, accy, accz, gyrox, gyroy, gyroz, magx, magy, magz, pos_prev, vel_prev, C_prev, acc_prev, P_prev, zupt):

    #use global dt and dgt variable in function
    global dt, dgt

    g = np.array([[0],[0],[9.8]])

    r_noise = 0.001
    acc_noise = 0.1
    gyr_noise = 0.01
    H = np.block([np.zeros((3,3)), np.zeros((3,3)), np.eye(3)])
    R = np.diag([r_noise**2,r_noise**2,r_noise**2])

    gy = np.deg2rad(gyroy)
    gx = np.deg2rad(gyrox)
    gz = np.deg2rad(gyroz)

    # Strapdown INS system
    ang_rate = np.array([[0, -gz, gy],
                         [gz, 0, -gx],
                         [-gy, gx, 0]])
    
    update_fac = (2*np.eye(3) + ang_rate*dgt) @ np.linalg.inv(2*np.eye(3) - ang_rate*dgt)
    C = C_prev @ update_fac

    acc_b = np.array([[accx],
                      [accy],
                      [accz]])

    acc_n = (0.5*(C+C_prev) @ acc_b)
    vel = (vel_prev + (acc_n + acc_prev -2*g).T * dt * 0.5).flatten()
    pos = pos_prev + (vel+vel_prev) * dt * 0.5

    S = np.array([[0, -acc_n[2,0], acc_n[1,0]],
                  [acc_n[2,0], 0, -acc_n[0,0]],
                  [-acc_n[1,0], acc_n[0,0], 0]])
                  
    F = np.block([[np.eye(3)      , np.zeros((3,3)), np.zeros((3,3))],
                  [np.zeros((3,3)), np.eye(3)      , np.eye(3)*dt],
                  [-S*dt          , np.zeros((3,3)), np.eye(3)]])
    
    Q = (np.diag([gyr_noise, gyr_noise, gyr_noise, 0, 0, 0, acc_noise, acc_noise, acc_noise])*dt) **2
    
    P = F @ P_prev @ F.T + Q

    if zupt:
        K = P @ H.T @ np.linalg.inv(H @ P @ H.T + R)
        error = K@vel

        P = (np.eye(9) - K @ H) @ P
        ang_rate_correction = np.array([[0, error[2], -error[1]],
                                        [-error[2], 0, error[0]],
                                        [error[1], -error[0], 0]])

        correction_factor = (2*np.eye(3) + ang_rate_correction) @ np.linalg.inv(2*np.eye(3) - ang_rate_correction)
        C = correction_factor@C        

        pos = pos - error[3:6]
        vel = vel - error[6:9]
    return pos, vel, C, acc_n, P
```

```python
def switch_case(velocity):
    vel_threshold = 1e-4
    if abs(velocity[0]) < vel_threshold and abs(velocity[1]) < vel_threshold:
        return 1    # trust ftm
    else:
        return 0    # trust pdr
```
